```python
import pandas as pd
from pycaret.datasets import get_data
from pycaret.regression import RegressionExperiment
```

```python
data = pd.read_pickle('df_init_time.pkl')
data.head()
```

```python
s = RegressionExperiment()
s.setup(data, target = 'df/f', session_id = 123)
```

```python
best = s.compare_models()
```

```python
s.evaluate_model(best)
```

```python
predicted = s.predict_model(best)
```
